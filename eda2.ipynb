### Introduction to the Exploratory Data Analysis (EDA) Project

In this project, we aim to perform **Exploratory Data Analysis (EDA)** on a dataset related to a marketing campaign, with the goal of understanding key insights and trends that can inform business decisions. The dataset includes various features that are critical to understanding customer behavior, campaign performance, and revenue generation. These features include campaign details, product information, customer demographics, and sales performance metrics.

The dataset contains the following columns:

1. **Campaign_ID**: Identifier for the specific marketing campaign.
2. **Product_ID**: Identifier for the product involved in the campaign.
3. **Budget**: The amount of money allocated to the campaign.
4. **Clicks**: The number of times the campaign ad was clicked.
5. **Conversions**: The number of successful actions (e.g., purchases) resulting from the campaign.
6. **Revenue_Generated**: The total revenue generated as a result of the campaign.
7. **ROI (Return on Investment)**: A key performance indicator that measures the profitability of the campaign.
8. **Customer_ID**: Unique identifier for customers who interacted with the campaign.
9. **Subscription_Tier**: The subscription level of the customer (e.g., Basic, Premium, VIP).
10. **Subscription_Length**: Duration of the customer's subscription to the service.
11. **Flash_Sale_ID**: Identifier for flash sales that occurred during the campaign.
12. **Discount_Level**: The level of discount applied during the campaign.
13. **Units_Sold**: The number of product units sold during the campaign.
14. **Bundle_ID**: Identifier for product bundles offered in the campaign.
15. **Bundle_Price**: Price of the product bundle.
16. **Customer_Satisfaction_Post_Refund**: Customer satisfaction rating after refund requests.
17. **Common_Keywords**: Keywords associated with the campaign, which may indicate themes or target audiences.

### Objective of the Analysis

The primary goal of this **Exploratory Data Analysis (EDA)** is to uncover valuable insights regarding the effectiveness of marketing campaigns, customer behavior, and factors that influence sales. Specifically, we will:

1. **Understand the distribution** of key metrics such as **Budget**, **Revenue**, **Units Sold**, and **ROI** to identify patterns in campaign performance.
2. **Analyze the relationship** between various campaign features (such as **Discount Level**, **Bundle Price**, and **Customer Subscription Tier**) and sales outcomes (like **Units Sold** and **Revenue Generated**).
3. **Investigate the influence** of customer-related factors, including **Subscription Length** and **Customer Satisfaction Post Refund**, on campaign success.
4. **Identify trends** and correlations that can guide future marketing strategies and resource allocation.

### Approach

To achieve these objectives, we will:
- **Clean and preprocess** the data to handle missing values, outliers, and data inconsistencies.
- Use **descriptive statistics** and **visualizations** (histograms, scatter plots, box plots) to summarize and understand key features.
- Investigate **correlations** between different features to find relationships that could indicate important patterns or dependencies.
- Conduct deeper analysis into specific variables (such as **Bundle Price** and **Discount Level**) to understand their direct impact on **Units Sold** and **Revenue Generated**.

By the end of this project, we aim to uncover insights that can inform decision-making in future campaigns, optimize resource allocation, and improve customer targeting.

```python
#1.What is the distribution of the `Budget` across campaigns?**
   #- This will help you understand how the marketing budget is spread out among campaigns.
# 1. Check summary statistics of 'Budget'
budget_summary = df['Budget'].describe()
print(budget_summary)

# 2. Plot Histogram for the 'Budget' distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Budget'], bins=20, kde=True)  # kde=True adds a smooth line (Kernel Density Estimate)
plt.title('Distribution of Budget across Campaigns')
plt.xlabel('Budget')
plt.ylabel('Frequency')
plt.show()

# 3. Boxplot to visualize the spread and outliers in 'Budget'
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Budget'])
plt.title('Boxplot of Budget across Campaigns')
plt.xlabel('Budget')
plt.show()

# 4. Identify outliers using IQR method
Q1 = df['Budget'].quantile(0.25)
Q3 = df['Budget'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find the outliers
outliers_budget = df[(df['Budget'] < lower_bound) | (df['Budget'] > upper_bound)]
print(f"Outliers in Budget:\n{outliers_budget}")
```

```python
#2. What is the distribution of `Clicks` and `Conversions` for each campaign?**
   # Analyzing this can reveal whether campaigns are effectively converting clicks into sales.
 #1. Check summary statistics for 'Clicks' and 'Conversions'
clicks_summary = df['Clicks'].describe()
conversions_summary = df['Conversions'].describe()

print(f"Summary of Clicks:\n{clicks_summary}")
print(f"\nSummary of Conversions:\n{conversions_summary}")

# 2. Plot Histograms for 'Clicks' and 'Conversions'
plt.figure(figsize=(14, 6))

# Plot Clicks distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Clicks'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Clicks across Campaigns')
plt.xlabel('Clicks')
plt.ylabel('Frequency')

# Plot Conversions distribution
plt.subplot(1, 2, 2)
sns.histplot(df['Conversions'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Conversions across Campaigns')
plt.xlabel('Conversions')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# 3. Plot Boxplots to visualize the spread and potential outliers for 'Clicks' and 'Conversions'
plt.figure(figsize=(14, 6))

# Boxplot for Clicks
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Clicks'], color='skyblue')
plt.title('Boxplot of Clicks across Campaigns')
plt.xlabel('Clicks')

# Boxplot for Conversions
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Conversions'], color='salmon')
plt.title('Boxplot of Conversions across Campaigns')
plt.xlabel('Conversions')

plt.tight_layout()
plt.show()

# 4. Investigate the relationship between 'Clicks' and 'Conversions'
# Scatter plot to observe the correlation between Clicks and Conversions
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Clicks', y='Conversions', color='purple')
plt.title('Clicks vs Conversions across Campaigns')
plt.xlabel('Clicks')
plt.ylabel('Conversions')
plt.show()

# 5. Correlation between Clicks and Conversions
correlation = df[['Clicks', 'Conversions']].corr()
print(f"\nCorrelation between Clicks and Conversions:\n{correlation}")
```

```python
#3.What is the average `ROI` (Return on Investment) for each `Campaign_ID`?**
   #is helps identify which campaigns provide the highest return relative to their budget.
# 1. Calculate ROI for each row (if not already provided in the dataset)
# ROI is calculated as (Revenue Generated - Budget) / Budget
df['Calculated_ROI'] = (df['Revenue_Generated'] - df['Budget']) / df['Budget']

# 2. Group by Campaign_ID and calculate the average ROI
average_roi_per_campaign = df.groupby('Campaign_ID')['Calculated_ROI'].mean()

# 3. Display the results
print(average_roi_per_campaign)
```

```python
#4.Which campaigns have the highest `Revenue_Generated` and why?**
   # Identifying high-revenue campaigns will help you analyze which strategies or products are most profitable.
 #1. Identify the campaigns with the highest 'Revenue_Generated'
top_revenue_campaigns = df.groupby('Campaign_ID')['Revenue_Generated'].sum().sort_values(ascending=False)

# 2. Display the top campaigns with the highest revenue
top_revenue_campaigns = top_revenue_campaigns.head()  # You can adjust the number to display more or fewer campaigns
print(top_revenue_campaigns)

# 3. Investigate potential reasons for high revenue
# For example, you can explore the distribution of factors like 'Discount_Level', 'Product_ID', 'Units_Sold', etc.
# Let's look at the campaign details that generated the highest revenue.

# Merge top campaigns with the original dataset to get detailed information
high_revenue_campaigns = df[df['Campaign_ID'].isin(top_revenue_campaigns.index)]

# Display detailed information for the high-revenue campaigns
print(high_revenue_campaigns[['Campaign_ID', 'Revenue_Generated', 'Discount_Level', 'Units_Sold', 'Product_ID', 'Bundle_Price']])
```

```python

#5. **What is the relationship between `Budget` and `Revenue_Generated`?**
   # This will help identify if higher budgets lead to higher revenues, or if other factors are at play.
# 1. Calculate the correlation between 'Budget' and 'Revenue_Generated'
correlation = df[['Budget', 'Revenue_Generated']].corr()

# Display the correlation matrix
print(f"Correlation between Budget and Revenue Generated:\n{correlation}")

# 2. Scatter plot to visualize the relationship between 'Budget' and 'Revenue_Generated'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Budget', y='Revenue_Generated', color='red')
plt.title('Relationship between Budget and Revenue Generated')
plt.xlabel('Budget')
plt.ylabel('Revenue Generated')
plt.show()

# 3. Plot a line of best fit to analyze the trend (optional)
plt.figure(figsize=(8, 6))
sns.regplot(data=df, x='Budget', y='Revenue_Generated', scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Trend of Budget vs Revenue Generated')
plt.xlabel('Budget')
plt.ylabel('Revenue Generated')
plt.show()
```

```python
#6. **What is the correlation between `Clicks` and `Conversions`?**
   # This can show how well a campaign is performing in terms of turning clicks into actual sales
# 1. Calculate the correlation between 'Clicks' and 'Conversions'
correlation = df[['Clicks', 'Conversions']].corr()

# Display the correlation matrix
print(f"Correlation between Clicks and Conversions:\n{correlation}")

# 2. Scatter plot to visualize the relationship between 'Clicks' and 'Conversions'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Clicks', y='Conversions', color='green')
plt.title('Relationship between Clicks and Conversions')
plt.xlabel('Clicks')
plt.ylabel('Conversions')
plt.show()

# 3. Plot a line of best fit to analyze the trend (optional)
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Clicks', y='Conversions', scatter_kws={'color': 'green'}, line_kws={'color': 'red'})
plt.title('Trend of Clicks vs Conversions')
plt.xlabel('Clicks')
plt.ylabel('Conversions')
plt.show()
```

```python

#7. **How does `Customer_Satisfaction_Post_Refund` vary across different `Subscription_Tiers`?**
 #  - This question helps analyze customer satisfaction levels based on the subscription tier they are in.
# 1. Group by 'Subscription_Tier' and calculate the mean satisfaction for each tier
satisfaction_by_tier = df.groupby('Subscription_Tier')['Customer_Satisfaction_Post_Refund'].mean()

# 2. Display the mean satisfaction for each subscription tier
print(satisfaction_by_tier)

# 3. To get a more detailed view, you can also calculate other summary statistics like median, min, max, etc.
satisfaction_summary = df.groupby('Subscription_Tier')['Customer_Satisfaction_Post_Refund'].describe()

# 4. Display the summary statistics
print(satisfaction_summary)
# 5. (Optional) Visualize the distribution of 'Customer_Satisfaction_Post_Refund' across subscription tiers using boxplot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Subscription_Tier', y='Customer_Satisfaction_Post_Refund')
plt.title('Customer Satisfaction Post Refund across Subscription Tiers')
plt.xlabel('Subscription Tier')
plt.ylabel('Customer Satisfaction Post Refund')
plt.show()
```

```python
#8. **How do `Units_Sold` correlate with `Bundle_Price` and `Discount_Level`?**
   #Analyzing these relationships could show whether offering discounts or bundling products impacts sales volume.
# 1. Calculate the correlation between 'Units_Sold', 'Bundle_Price', and 'Discount_Level'
correlation_matrix = df[['Units_Sold', 'Bundle_Price', 'Discount_Level']].corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# 2. Visualize the relationships with scatter plots

# Scatter plot for Units_Sold vs Bundle_Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Bundle_Price', y='Units_Sold', color='blue')
plt.title('Units Sold vs Bundle Price')
plt.xlabel('Bundle Price')
plt.ylabel('Units Sold')
plt.show()

# Scatter plot for Units_Sold vs Discount_Level
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount_Level', y='Units_Sold', color='green')
plt.title('Units Sold vs Discount Level')
plt.xlabel('Discount Level')
plt.ylabel('Units Sold')
plt.show()
```

```python
#9. **What is the average `Subscription_Length` for each `Subscription_Tier`?**
   #This gives insights into how long customers typically stay with different subscription plans.
# 1. Group by 'Subscription_Tier' and calculate the mean of 'Subscription_Length'
average_subscription_length = df.groupby('Subscription_Tier')['Subscription_Length'].mean()

# 2. Display the average subscription length for each subscription tier
print("Average Subscription Length for each Subscription Tier:")
print(average_subscription_length)
```

```python
#10. **How does the `Discount_Level` affect the `Units_Sold`?**
    # This helps determine if higher discounts lead to more sales, or if there's a point where discounts stop being effective.
# 1. Calculate the correlation between 'Discount_Level' and 'Units_Sold'
correlation = df[['Discount_Level', 'Units_Sold']].corr()

# Display the correlation
print(f"Correlation between Discount Level and Units Sold:\n{correlation}")

# 2. Visualize the relationship with a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount_Level', y='Units_Sold', color='blue')
plt.title('Units Sold vs Discount Level')
plt.xlabel('Discount Level')
plt.ylabel('Units Sold')
plt.show()
 #3Group by 'Discount_Level' and calculate the average 'Units_Sold' for each discount level
average_units_sold_by_discount = df.groupby('Discount_Level')['Units_Sold'].mean()

# Display the average units sold for each discount level
print(f"Average Units Sold for each Discount Level:\n{average_units_sold_by_discount}")
```

```python
#11.What is the impact of `Flash_Sale_ID` on `Revenue_Generated` and `Units_Sold`?**
    #This question can help determine the success of flash sales in terms of both sales and revenue.

# 1. Group by 'Flash_Sale_ID' and calculate the total 'Revenue_Generated' and 'Units_Sold'
flash_sale_impact = df.groupby('Flash_Sale_ID')[['Revenue_Generated', 'Units_Sold']].sum()

# 2. Display the total revenue and units sold for each flash sale
print("Impact of Flash Sale on Revenue Generated and Units Sold:")
print(flash_sale_impact)
# 5. (Optional) Calculate the average revenue and units sold per flash sale
average_flash_sale_impact = df.groupby('Flash_Sale_ID')[['Revenue_Generated', 'Units_Sold']].mean()

# Display the average revenue and units sold per flash sale
print("Average Revenue and Units Sold per Flash Sale:")
print(average_flash_sale_impact)
```

```python
#12.Is there a correlation between customer satisfaction and units sold?
# Calculate the correlation between Customer Satisfaction and Units Sold
correlation = df['Customer_Satisfaction_Post_Refund'].corr(df['Units_Sold'])

# Print the correlation value
print(f'Correlation between Customer Satisfaction and Units Sold: {correlation}')
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Customer_Satisfaction_Post_Refund', y='Units_Sold', data=df)

# Set plot title and labels
plt.title('Relationship between Customer Satisfaction and Units Sold')
plt.xlabel('Customer Satisfaction Post Refund')
plt.ylabel('Units Sold')
plt.show()
```

```python
#13.What is the average revenue generated per customer based on their subscription tier?
# Group the data by Subscription_Tier and calculate the average revenue per customer
average_revenue_per_tier = df.groupby('Subscription_Tier')['Revenue_Generated'].mean()

# Print the results
print(average_revenue_per_tier)

# Plot the average revenue per subscription tier
plt.figure(figsize=(8, 6))
average_revenue_per_tier.plot(kind='bar', color='skyblue')

# Set title and labels
plt.title('Average Revenue Generated per Customer by Subscription Tier')
plt.xlabel('Subscription Tier')
plt.ylabel('Average Revenue Generated')
plt.xticks(rotation=45)
plt.show()
```

```python
#14.What is the overall Return on Investment (ROI) for each product in the dataset?
# Group the data by Product_ID and calculate the total revenue and budget for each product
roi_data = df.groupby('Product_ID').agg(
    total_revenue=('Revenue_Generated', 'sum'),
    total_budget=('Budget', 'sum')
).reset_index()

# Calculate the ROI for each product
roi_data['ROI'] = ((roi_data['total_revenue'] - roi_data['total_budget']) / roi_data['total_budget']) * 100

# Display the ROI for each product
print(roi_data[['Product_ID', 'ROI']])

```

```python
#15.What is the impact of different bundle prices on the number of units sold?
# Display first few rows of the dataset
print(df[['Bundle_Price', 'Units_Sold']].head())
# Group by Bundle_Price and calculate the average units sold
bundle_price_impact = df.groupby('Bundle_Price')['Units_Sold'].mean().reset_index()

# Sort the data to see the impact
bundle_price_impact = bundle_price_impact.sort_values(by='Bundle_Price', ascending=True)

print(bundle_price_impact)
# Create a scatter plot to see the relationship between Bundle Price and Units Sold
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Bundle_Price', y='Units_Sold', data=df, color='b')

# Add title and labels
plt.title('Impact of Bundle Price on Units Sold')
plt.xlabel('Bundle Price')
plt.ylabel('Units Sold')
plt.show()
```
